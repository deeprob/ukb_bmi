```python
import pandas as pd
```

```python
combo2_file = "/data5/deepro/ukbiobank/papers/bmi_project/3_run_rarecomb/white_british/data/parsed_tables/combo_2.csv"
combo3_file = "/data5/deepro/ukbiobank/papers/bmi_project/3_run_rarecomb/white_british/data/parsed_tables/combo_3.csv"
combo2_lfr_file = "/data5/deepro/ukbiobank/papers/bmi_project/3_run_rarecomb/white_british_lifestyle_regressed/data/parsed_tables/combo_2.csv"
combo3_lfr_file = "/data5/deepro/ukbiobank/papers/bmi_project/3_run_rarecomb/white_british_lifestyle_regressed/data/parsed_tables/combo_3.csv"
```

```python
combo2_df = pd.read_csv(combo2_file)
combo3_df = pd.read_csv(combo3_file)
combo2_lfr_df = pd.read_csv(combo2_lfr_file)
combo3_lfr_df = pd.read_csv(combo3_lfr_file)
```

```python
def get_sorted_combos(ser, ncombo):
    return ";".join(sorted(ser.loc[[f"Item_{i}_symbol" for i in range(1, ncombo + 1)]]))
```

```python
combo2_df["combo_name"] = combo2_df.apply(get_sorted_combos, args=(2, ), axis=1)
combo3_df["combo_name"] = combo3_df.apply(get_sorted_combos, args=(3, ), axis=1)
combo2_lfr_df["combo_name"] = combo2_lfr_df.apply(get_sorted_combos, args=(2, ), axis=1)
combo3_lfr_df["combo_name"] = combo3_lfr_df.apply(get_sorted_combos, args=(3, ), axis=1)
```

```python
og_set = set(combo2_df.combo_name.to_list() + combo3_df.combo_name.to_list())
lfr_set = set(combo2_lfr_df.combo_name.to_list() + combo3_lfr_df.combo_name.to_list())
```

```python
len(og_set), len(lfr_set)
```

```python
int_set = og_set.intersection(lfr_set)
```
