```python
import os
import pandas as pd
from functools import reduce
```

```python
combo_info_df = pd.read_csv("/data6/deepro/ukb_bmi/3_characterization/data/combos_in_cohort/protective/test_cohort.csv")

combo_files = [
    "/data6/deepro/ukb_bmi/2_rarecomb/data/protective/combo2.csv",
    "/data6/deepro/ukb_bmi/2_rarecomb/data/protective/combo3.csv"
]
```

```python
combo_df = pd.concat([pd.read_csv(cf, usecols=["uniq_items", "Case_Samples", "Control_Samples"]) for cf in combo_files])
```

```python
combo_info_df = combo_info_df.merge(combo_df, on="uniq_items")
```

```python
combo_genes = set("|".join(combo_info_df.uniq_items.str.replace("Input_", "").values).split("|"))
combo_samples = set("|".join(combo_info_df.combo_samples.fillna("").values).split("|"))
combo_case_samples = set("|".join(combo_info_df.Case_Samples.values).split("|"))
```

```python
def get_vtype_df(block_df, combo_genes, combo_samples, combo_case_samples):
    # select lof and deleterious missense variants
    block_df = block_df.loc[(block_df.lof==True)|(block_df.splice_lof==True)|((block_df.missense==True)&(block_df.del_score>4))]
    block_df["samples"] = block_df.samples.str.split(",")
    block_df = block_df.explode("samples")
    block_df = block_df.loc[(block_df.samples.isin(combo_samples.union(combo_case_samples))) & (block_df.gene.isin(combo_genes))]
    return block_df
```

```python
vcfs_per_chrm = {
    "chr1": 97, "chr2": 71, "chr3": 56, "chr4": 39, "chr5": 43, "chr6": 48, 
    "chr7": 47, "chr8": 35, "chr9": 42, "chr10": 40, "chr11": 57, "chr12": 52, 
    "chr13": 18, "chr14": 30, "chr15": 34, "chr16": 47, "chr17": 56, "chr18": 16, 
    "chr19": 65, "chr20": 25, "chr21": 11, "chr22": 23, "chrX": 24, "chrY": 1
}
annot_table_dir = "/data6/deepro/ukb_bmi/0_data_preparation_and_download/exome_annot/data/annot_tables_vep109"

cohort_df = pd.DataFrame()
for chr_num in [f"chr{i}" for i in range(1,23)] + ["chrX", "chrY"]:
    print(chr_num)
    chr_file_num = vcfs_per_chrm[chr_num]
    for filei in range(chr_file_num):
        block_file = os.path.join(annot_table_dir, f"{chr_num}", f"block_{filei}.tsv.gz")
        block_df = pd.read_csv(block_file, sep="\t", index_col=0)
        block_df = get_vtype_df(block_df, combo_genes, combo_samples, combo_case_samples)
        cohort_df = pd.concat((cohort_df, block_df))
```

```python
combo_info_df = combo_info_df.fillna("")
```

```python
def get_samples_with_same_variants(ser, cohort_df, vtype="locus"):
    genes = ser.uniq_items.replace("Input_", "").split("|")
    case_samples = set(ser.Case_Samples.split("|"))
    all_samples = set(ser.combo_samples.split("|"))
    other_samples = all_samples.difference(case_samples)
    cohort_case_df = cohort_df.loc[cohort_df.samples.isin(case_samples)]
    variants_by_gene = [set(cohort_case_df.loc[cohort_case_df.gene==g, vtype].values) for g in genes]
    cohort_other_df = cohort_df.loc[cohort_df.samples.isin(other_samples)]
    # get gene wise cohort samples who contain at least one of the case sample variants
    cohort_samples_w_similar_variants = [set(cohort_other_df.loc[(cohort_other_df.gene==g) & (cohort_other_df[vtype].isin(v)), "samples"].values) for g,v in zip(genes, variants_by_gene)]
    cohort_samples_w_same_variants = reduce(lambda x,y: x.intersection(y), cohort_samples_w_similar_variants)
    cohort_samples_w_different_variants = other_samples.difference(cohort_samples_w_same_variants)
    return pd.Series({"cohort_samples_same_variants": "|".join(cohort_samples_w_same_variants), "cohort_samples_other_variants": "|".join(cohort_samples_w_different_variants)})
```

```python
combo_info_df = combo_info_df.merge(combo_info_df.apply(get_samples_with_same_variants, args=(cohort_df, "locus"), axis=1), left_index=True, right_index=True)
```
